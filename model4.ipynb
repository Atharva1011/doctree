```python
import re
def getHostName(url):
    match = re.match(r"^(?:https?:\/\/)?(?:[^@\n]+@)?(?:www\.)?([^:\/\n]+)", url)
    if match:
        host_name = match.group(1)
        return host_name
```

```python
train_df["HOSTNAME"] = train_df["URL"].apply(getHostName)
```
